```python
# 40
class Morph:
    """
    def __init__(self, surface, base, pos, pos1):
        self.surface = surface
        self.base = base
        self.pos = pos
        self.pos1 = pos1
    """
    def __init__(self, morph):
        v = morph["feature"].split(",")
        self.surface = morph.text
        self.base = v[6]
        self.pos = v[0]
        self.pos1 = v[1]
    def __str__(self):
        return '(表層形：%s, 基本形：%s, 品詞：%s, 品詞細分類: %s)' % (self.surface, self.base, self.pos, self.pos1)
    
    def __repr__(self):
        return str(self)
        
data = []
parsed = Path("ai.ja.txt.parsed").read_text()
soup = BeautifulSoup(parsed, "html.parser")
for chunk in soup.find_all("chunk"):
    for morph in chunk.find_all("tok"):
        # v = morph["feature"].split(",")
        # data.append(Morph(morph.text, v[6], v[0], v[1]))
        data.append(Morph(morph))

for d in data:
    print(d)
    if d.surface == '。':
        break
print(data)
```

```python
# 41
class Chunk():
    def __init__(self, morphs, dst, srcs):
        self.morphs = morphs
        self.dst = dst
        self.srcs = srcs
    """
    def __init__(self, sentence):
        self.morphs = morphs
        self.dst = dst
        self.srcs = srcs
    """
    def __str__(self):
        return '形態素：%s, 係受け先：%s, 係受け元：%s' % (self.morphs, self.dst, self.srcs)

    def __repr__(self):
        return str(self)
        
def get_chunks_list(filename="ai.ja.txt.parsed"):
    parsed = Path(filename).read_text()
    soup = BeautifulSoup(parsed, "html.parser")
    chunks_list = []
    for sentence in soup.find_all("sentence"):
        found_chunks = sentence.find_all("chunk")
        if len(found_chunks) == 0:
            continue
        start_ind = int(found_chunks[0]["id"])
        linked = [ [] for _ in range(len(found_chunks))]
        for i, chunk in enumerate(found_chunks):
            dst = int(chunk["link"]) - start_ind
            if dst == -1:
                continue
            linked[dst].append(i)
        chunks = []
        for i, chunk in enumerate(found_chunks):
            morphs = []
            for morph in chunk.find_all("tok"):
                morphs.append(Morph(morph))
            dst = int(chunk["link"]) - start_ind
            chunks.append(Chunk(morphs, dst, linked[i]))
        chunks_list.append(chunks)
    return chunks_list
    
chunks_list = get_chunks_list()
print(chunks_list[0])
```

```python
# 42
import re
code_regex = re.compile('[!"#$%&\'\\\\()*+,-./:;<=>?@[\\]^_`{|}~「」〔〕“”〈〉『』【】＆＊・（）＄＃＠。、？！｀＋￥％]')

def show_relation(chunks):
    surfaces = []
    for chunk in chunks:
        surface = ""
        for morph in chunk.morphs:
            surface += morph.surface
        surfaces.append(code_regex.sub('', surface))
    for i, chunk in enumerate(chunks):
        str = f'"{surfaces[i]}"\n\t係り受け先: '
        if chunk.dst == -1:
            str += "なし"
        else:
            str += surfaces[chunk.dst]
        str += "\n\t係り受け元: "
        for i in chunk.srcs:
            str += f'{surfaces[i]}\t'
        print(str)
        
for chunks in chunks_list:
    show_relation(chunks)
```

```python
# 43
import re
code_regex = re.compile('[!"#$%&\'\\\\()*+,-./:;<=>?@[\\]^_`{|}~「」〔〕“”〈〉『』【】＆＊・（）＄＃＠。、？！｀＋￥％]')

def show_relation_from_noun_to_verb(chunks):
    surfaces = []
    for chunk in chunks:
        surface = ""
        for morph in chunk.morphs:
            surface += morph.surface
        surfaces.append(code_regex.sub('', surface))
    for i, chunk in enumerate(chunks):
        include_noun = any([ morph.pos == '名詞' for morph in chunk.morphs ])
        if not include_noun:
            continue
        if chunk.dst == -1:
            continue
        include_verb = any([ morph.pos == '動詞' for morph in chunks[chunk.dst].morphs ])
        if not include_verb:
            continue
        print(surfaces[i] + "\t" + surfaces[chunk.dst])
        
for chunks in chunks_list:
    show_relation_from_noun_to_verb(chunks)
```

```python
# 45
# https://yucatio.hatenablog.com/entry/2019/12/18/214251
DAKU_TO_SEION = {
    'ゔ': 'う',
    'が': 'が', 'ぎ': 'き', 'ぐ': 'く', 'げ': 'け', 'ご': 'こ',
    'ざ': 'さ', 'じ': 'し', 'ず': 'す', 'ぜ': 'せ', 'ぞ': 'そ',
    'だ': 'た', 'ぢ': 'ち', 'づ': 'つ', 'で': 'て', 'ど': 'と',
    'ば': 'は', 'び': 'ひ', 'ぶ': 'ふ', 'べ': 'へ', 'ぼ': 'ほ',
    'ぱ': 'は', 'ぴ': 'ひ', 'ぷ': 'ふ', 'ぺ': 'へ', 'ぽ': 'ほ',
}

SMALL_TO_NORMAL = {
    'ぁ': 'あ', 'ぃ': 'い', 'ぅ': 'う', 'ぇ': 'え', 'ぉ': 'お',
    'っ': 'つ',
    'ゃ': 'や', 'ゅ': 'ゆ', 'ょ': 'よ',
}

TO_NORMAL_SEION_TABLE = str.maketrans({**DAKU_TO_SEION, **SMALL_TO_NORMAL})

TO_NORMAL_TABLE = str.maketrans(SMALL_TO_NORMAL)


def sort_as_ja_dict(kana_list, *, key=lambda x: x):
    return sorted(kana_list,
                  key=lambda kana: (key(kana).translate(TO_NORMAL_SEION_TABLE),
                                    key(kana).translate(TO_NORMAL_TABLE), kana))
```

```python
# 45
import re
code_regex = re.compile('[!"#$%&\'\\\\()*+,-./:;<=>?@[\\]^_`{|}~「」〔〕“”〈〉『』【】＆＊・（）＄＃＠。、？！｀＋￥％]')

chunks_list = get_chunks_list()
for chunks in chunks_list:
    for chunk in chunks:
        for morph in chunk.morphs:
            morph.surface = code_regex.sub('', morph.surface)

output_str = ""
for chunks in chunks_list:
    for i, chunk in enumerate(chunks):
        jutsugo = next((morph for morph in chunk.morphs if morph.pos == '動詞'), None)
        if jutsugo == None:
            continue
        kaku = list(map(lambda j: chunks[j].morphs, chunk.srcs))
        for j, k in enumerate(kaku):
            k = list(filter(lambda m: m.pos == '助詞', k))
            k = list(map(lambda m: m.surface, k))
            kaku[j] = k
        kaku = sum(kaku, [])
        kaku = sort_as_ja_dict(kaku)
        output_str += f'{jutsugo.base} {" ".join(kaku)}\n'
        if jutsugo.base == '作り出す':
            print(f'{jutsugo.base} {" ".join(kaku)}')
with open('tmp/corpus-45.txt', 'w') as f:
    f.write(output_str)
```

```python
# 47
import re
code_regex = re.compile('[!"#$%&\'\\\\()*+,-./:;<=>?@[\\]^_`{|}~「」〔〕“”〈〉『』【】＆＊・（）＄＃＠。、？！｀＋￥％]')

chunks_list = get_chunks_list()
for chunks in chunks_list:
    for chunk in chunks:
        for morph in chunk.morphs:
            morph.surface = code_regex.sub('', morph.surface)

for chunks in chunks_list:
    for id, chunk in enumerate(chunks):
        morphs = chunk.morphs
        if len(morphs) <= 1:
            continue
        all_sahen = True
        jutsugo = ""
        for i in range(len(morphs) - 1):
            if morphs[i].pos != '名詞' or morphs[i].pos1 != 'サ変接続':
                all_sahen = False
            jutsugo += morphs[i].surface
        josi = morphs[len(morphs) - 1]
        if not all_sahen or josi.pos != '助詞' or josi.surface != 'を':
            continue
        jutsugo += 'を'
        if chunk.dst == -1:
            continue
        dousi_morphs = chunks[chunk.dst].morphs
        if len(dousi_morphs) != 1 or dousi_morphs[0].pos != '動詞':
            continue
        jutsugo += dousi_morphs[0].surface
        srcs = chunk.srcs + chunks[chunk.dst].srcs
        srcs = list(set(srcs))
        srcs = list(filter(lambda x: x != id and x != chunk.dst, srcs))
        morphs_list = list(map(lambda j: chunks[j].morphs, srcs))
        morph_tuple_list = []
        for morphs in morphs_list:
            surface = ""
            for m in morphs:
                surface += m.surface
                if m.pos == '助詞':
                    morph_tuple_list.append((surface, m.surface))
                    surface = ""
                    continue
        if len(morph_tuple_list) == 0:
            continue
        morph_tuple_list = sort_as_ja_dict(morph_tuple_list, key=lambda x: x[1])
        kaku = []
        kou = []
        for x in morph_tuple_list:
            kaku.append(x[1])
            kou.append(x[0])
            
        #if jutsugo == '学習を行う':
        print(f'{jutsugo} {" ".join(kaku)} {" ".join(kou)}')
```

```python
# 49
# ! pip install cabocha-python
import re
code_regex = re.compile('[!"#$%&\'\\\\()*+,-./:;<=>?@[\\]^_`{|}~「」〔〕“”〈〉『』【】＆＊・（）＄＃＠。、？！｀＋￥％]')
            
# 名詞句は名詞を含む文節とする
n = len(chunks)
def get_list(nxt):
    ret = []
    while nxt != -1:
        ret.append(nxt)
        nxt = chunks[nxt].dst
    return ret
    
def get_relation(i, j):
    i_list = get_list(i)
    j_list = get_list(j)
    li = len(i_list)
    lj = len(j_list)
    ret = []
    common = []
    for idx, di in enumerate(i_list):
        if di in j_list:
            k = j_list.index(di)
            l = min(li - idx, lj - k)
            for dl in range(l):
                if i_list[idx + dl] == j_list[k + dl]:
                    common.append(i_list[idx + dl])
                else:
                    break
            if len(common) == len(j_list):
                ret.append(i_list[:idx])
                ret.append(j_list[:1])
            else:
                ret.append(i_list[:idx])
                ret.append(j_list[:k])
                ret.append(common)
            return ret
    return None

chunks_list = get_chunks_list("cabocha-48-example.txt.parsed")
for chunks in chunks_list:
    for chunk in chunks:
        for morph in chunk.morphs:
            morph.surface = code_regex.sub('', morph.surface)
for chunks in chunks_list:
    n = len(chunks)
    for i in range(n):
        for j in range(i + 1, n):
            #print(get_list(i), get_list(j))
            meisiku_i = any([ morph.pos == '名詞' for morph in chunks[i].morphs ])
            meisiku_j = any([ morph.pos == '名詞' for morph in chunks[j].morphs ])
            # print(chunks[i].morphs, chunks[j].morphs)
            # print(meisiku_i, meisiku_j)
            if not meisiku_i or not meisiku_j:
                continue
            lst = get_relation(i, j)
            #print(lst)
            if len(lst) == 3:
                lstr = ""
                for is_end, id in enumerate(lst[0]):
                    for morph in chunks[id].morphs:
                        if is_end == 0 and morph.pos == '名詞':
                            if len(lstr) == 0:
                                lstr += 'X'
                            continue
                        lstr += morph.surface
                    if is_end + 1 < len(lst[0]):
                        lstr += " -> "
                mstr = "| "
                for is_end, id in enumerate(lst[1]):
                    for morph in chunks[id].morphs:
                        if is_end == 0 and morph.pos == '名詞':
                            if len(mstr) == 2:
                                mstr += 'Y'
                            continue
                        mstr += morph.surface
                    if is_end + 1 < len(lst[1]):
                        mstr += " -> "
                rstr = "| "
                for is_end, id in enumerate(lst[2]):
                    for morph in chunks[id].morphs:
                        rstr += morph.surface
                    if is_end + 1 < len(lst[2]):
                        rstr += " -> "
                print(lstr, mstr, rstr)
            elif len(lst) == 2:
                lstr = ""
                for is_end, id in enumerate(lst[0]):
                    for morph in chunks[id].morphs:
                        if is_end == 0 and morph.pos == '名詞':
                            if len(lstr) == 0:
                                lstr += 'X'
                            continue
                        lstr += morph.surface
                    if is_end + 1 < len(lst[0]):
                        lstr += " -> "
                mstr = "-> "
                for is_end, id in enumerate(lst[1]):
                    for morph in chunks[id].morphs:
                        if is_end == 0 and morph.pos == '名詞':
                            if len(mstr) == 3:
                                mstr += 'Y'
                            continue
                        mstr += morph.surface
                    if is_end + 1 < len(lst[1]):
                        mstr += " -> "
                print(lstr, mstr)
```
