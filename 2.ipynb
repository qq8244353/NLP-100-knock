```python
# 21
import re
categories = []
with open('tmp/article.txt', 'r') as f:
    for l in f:
        res = re.search(r'Category', l)
        if res != None:
            print(l, end="")
            categories.append(l[:-1])

categories = categories[1:]
```

```python
# 24
lst = []
with open('tmp/article.txt', 'r') as f:
    for l in f:
        res = re.search(r'ファイル', l)
        if res != None:
            lst.append(re.findall(r'ファイル:([^\|\]]*)', l))
lst = sum(lst, [])
for l in lst:
    print(l)
```

```python
# 25
def f25():
    map = {}
    with open('tmp/article.txt', 'r') as f:
        tmp = ""
        content = False
        for l in f:
            sta = re.search(r'{{基礎情報', l)
            if sta != None:
                content = True
                continue
            if content:
                fin = re.search(r'^}}', l)
                if fin != None:
                    k = re.findall(r'^\|(\S*)', tmp)[0]
                    v = re.findall(r'= {,1}(.*)', tmp)[0]
                    map[k] = v
                    break
            if content:
                start_or = re.search(r'^\|', l)
                if start_or != None:
                    if tmp == "":
                        tmp += l[:-1]
                        continue
                    k = re.findall(r'^\|(\S*)', tmp)[0]
                    v = re.findall(r'= {,1}(.*)', tmp)[0]
                    map[k] = v
                    tmp = ""
                    tmp += l[:-1]
                else:
                    tmp += l[:-1]
                    continue
    return map

m = f25()
for k, v in m.items():
    print(k, v)
```

```python
# 26
def f26():
    m = f25()
    for k, v in m.items():
        m[k] = re.sub(r'\'{,2}', '', v)
    return m

m = f26()
for k, v in m.items():
    print(k, v)
```

```python
# 27
def f27():
    m = f26()
    for k in m.keys():
        m[k] = re.sub(r'\[{2}[^\|]*?\|*([^\|\]]+)\]{2}', r'\1', m[k])
        m[k] = re.sub(r'\[{2}[^\{]*\|([^\]]+)\]{2}.*', r'\1', m[k])
    return m
    
m = f27()
for k, v in m.items():
    print(k, v)
```

```python
# 28
def f28():
    m = f27()
    for k in m.keys():
        m[k] = re.sub(r'<(?!references).*>', '', m[k])
        m[k] = re.sub(r'\{\{0\}\}', '', m[k])
        m[k] = re.sub(r'\{\{.*?([^\|]+)\}\}', r'\1', m[k])
        m[k] = re.sub(r'\(\+over.*', r'', m[k])
    return m
    
m = f28()
for k, v in m.items():
    print(k, v)
```

```python
# 29
import requests
def fetch_image_src(filename):
    S = requests.Session()
    params = {
        "action": "query",
        "format": "json",
        "prop": "imageinfo",
        "iiprop": "url",
        "titles": f'File:{filename}'
    }

    response = S.get(url = "https://en.wikipedia.org/w/api.php", params = params)
    data = response.json()
    page = next(iter(data["query"]["pages"].values()))
    image_info = page["imageinfo"][0]
    image_url = image_info["url"]

    return image_url
m = f28()

fetch_image_src(m['国旗画像'])
```
