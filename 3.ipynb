```python
# 30
import re
def load():
    ret = []
    with open('neko.txt.mecab') as f:
        content = f.readlines()
    for l in content:
        v = re.split(r'[\t,]', l[:-1])
        if len(v) != 10:
            continue
        ret.append({ 'surface' : v[0], 'base' : v[7], 'pos' : v[1], 'pos1' : v[2] })
    return ret
data = load()
data
```

```python
# 31
verbs = []
for d in data:
    if d['pos'] == '動詞':
        verbs.append(d['surface'])
verbs
```

```python
# 32
verb_bases = []
for d in data:
    if d['pos'] == '動詞':
        verb_bases.append(d['base'])
verb_bases
```

```python
# 33
n = len(data)
noun_no = []
for i in range(n - 2):
    if data[i]['pos'] == '名詞' and data[i + 1]['surface'] == 'の' and data[i + 2]['pos'] == '名詞':
        noun_no.append(data[i]['surface'] + data[i + 1]['surface'] + data[i + 2]['surface'])
noun_no
```

```python
# 34
noun_succ = []
i = 0
n = len(data)
while i < n:
    str = ""
    while i < n and data[i]['pos'] == '名詞':
        str += data[i]['surface']
        i += 1
    if len(str) > 0:
        noun_succ.append(str)
    i += 1 # data[i]は 参照済み or i >= n or 名詞ではない なので次に行ってよし
noun_succ
```

```python
# 35
# 単語
from collections import defaultdict
cnt = defaultdict(int)
freq_ranking = []
for d in data:
    if cnt[d['surface']] == 0:
        freq_ranking.append(d['surface'])
    cnt[d['surface']] += 1
freq_ranking = sorted(freq_ranking, key=lambda x: cnt[x], reverse=True)
freq_ranking
```

```python
# 37
now = []
cnt = defaultdict(int)
freq_ranking = []
for d in data:
    freq_ranking.append(d['surface'])
    now.append(d['surface'])
    if d['surface'] == '。':
        exist = False
        for n in now:
            if n == '猫':
                exist = True
        if not exist:
            now = []
            continue
        for n in now:
            if n == '猫':
                continue
            cnt[n] += 1
        now = []
freq_ranking = list(set(freq_ranking))
#print(freq_ranking)
freq_ranking = sorted(freq_ranking, key=lambda x: cnt[x], reverse=True)
#print(freq_ranking)
df = pd.DataFrame(list(map(lambda x: cnt[x], freq_ranking[:10])), index=freq_ranking[:10], columns=['frequency'])
df.plot.bar()
```

```python
# 38
freq_ranking = []
cnt = defaultdict(int)
for d in data:
    freq_ranking.append(d['surface'])
    cnt[d['surface']] += 1
freq_ranking = list(set(freq_ranking))
freq_ranking = sorted(freq_ranking, key=lambda x: cnt[x])

index = list(set(map(lambda x : cnt[x], freq_ranking)))
cnt_by_freq = defaultdict(int)
for f in freq_ranking:
    cnt_by_freq[cnt[f]] += 1
#print(cnt_by_freq)
#print(index)
df = pd.DataFrame(list(map(lambda x: cnt_by_freq[x], index)), index=index, columns=['種類数'])
df.plot.hist(bins=200)
```
